```python
import math
from matplotlib import pyplot as plt

x = list (range(0,100))
y1 = [math.pow(num,2) for num in x ]
y2 = [math.pow(num,3) for num in x ]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))

ax1.plot(x, y1 , colour = 'red')

ax2.plot(x, y2 , colour = 'blue')
```
